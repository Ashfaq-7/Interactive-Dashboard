```python
import hvplot.pandas
import holoviews as hv

hv.extension('bokeh')
```

```python
(food[
    (food.cook_time == 30) &
    (food.course.isin(['main course', 'snack', 'dessert']))
]
.groupby(['diet','course','flavor_profile','cook_time'])['prep_time'].mean()
.to_frame()
.reset_index()
.sort_values(by='cook_time')
).head(10)
```

```python
cook_time=pn.widgets.IntSlider(name='cook_time', start=30, end=120, step=5)
course=pn.widgets.ToggleGroup(name='course', options=['main course', 'snack', 'dessert'],
                          value=['main course', 'snack', 'dessert'],
button_type='warning')
#success, danger
yaxis=pn.widgets.RadioButtonGroup(name='Y axis', options=['prep_time', 'cook_time'], button_type='success')
```

```python
idf=food.interactive()
```

```python
ipipeline=(
idf[
    (idf.cook_time==cook_time) &
    (idf.course.isin(course))
]
    .groupby(['course', 'flavor_profile'])[yaxis].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='flavor_profile')
    .reset_index(drop=True)
)
ipipeline.head()
```

```python
scatter = ipipeline.hvplot.scatter(x='course', y='prep_time', c='flavor_profile', title='Course vs Prep_time')
scatter
```

```python
line = ipipeline.hvplot.line(x='course', y='prep_time', by='flavor_profile', title='Ingredients vs Prep_time (Line)')
line
```

```python
hist = ipipeline.hvplot.hist(y='prep_time', by='course', bins=20, title='Prep_time Distribution')
hist
```

```python
bar = ipipeline.groupby('flavor_profile').mean().hvplot.bar(y='prep_time', title='Average Prep_time per flavor_profile')
bar
```

```python
box = ipipeline.hvplot.box(y='prep_time', by='flavor_profile', title='Prep_time Distribution by flavor_profile')
box
```

```python
violin = ipipeline.hvplot.violin(y='prep_time', by='course', title='Prep time Distribution by Course (Violin Plot)')
violin
```

```python
heatmap = ipipeline.corr().hvplot.heatmap(title='Correlation Matrix', cmap='viridis', width=400, height=400)
heatmap
```
